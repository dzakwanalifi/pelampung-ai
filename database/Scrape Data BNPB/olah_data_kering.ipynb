```python
import pandas as pd
```

```python
# Load data from CSV
df = pd.read_csv('data_kering.csv')

# Convert DataFrame to Excel
df.to_excel('data_kering.xlsx', index=False)
```

```python
# Memuat data dari file CSV
data = pd.read_excel('data_kering.xlsx')
```

```python
# Memisahkan kolom "Detail" menjadi beberapa kolom baru
data[['Meninggal', 'Hilang', 'Terluka', 'Menderita', 'Mengungsi']] = data['Detail'].str.extract('Meninggal : (\d+)<br>Hilang : (\d+)<br>Terluka : (\d+)<br>Menderita : (\d+)<br>Mengungsi : (\d+)<br>')
data[['Rumah', 'Fas_Pendidikan', 'Fas_Kesehatan', 'Fas_Peribadatan', 'Fas_Umum', 'Perkantoran', 'Jembatan', 'Pabrik', 'Pertokoan']] = data['Detail'].str.extract('Rumah : (\d+)<br>Fas.Pendidikan : (\d+)<br>Fas.Kesehatan : (\d+)<br>Fas.Peribadatan : (\d+)<br>Fas.Umum : (\d+)<br>Perkantoran : (\d+)<br>Jembatan : (\d+)<br>Pabrik : (\d+)<br>Pertokoan : (\d+)<br>')
```

```python
# Menghapus kolom "Detail" yang tidak diperlukan lagi
data = data.drop(columns='Detail')
data = data.drop(columns='No')
```

```python
data.fillna(0, inplace=True)
```
